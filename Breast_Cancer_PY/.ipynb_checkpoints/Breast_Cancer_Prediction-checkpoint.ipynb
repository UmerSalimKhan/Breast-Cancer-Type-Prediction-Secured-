# Splitting dataset in training & testing data

```python
x_train, x_test, y_train, y_test = train_test_split (X, y, test_size=0.3, random_state=42)
print ("Shape \nX_train: {}, y_train: {}, x_test: {}, y_test: {} \n".format (x_train.shape, y_train.shape, x_test.shape, y_test.shape))
print ("Length \nX_train {}, y_train {}, X_test {}, y_test {}".format (len (x_train), len (y_train), len (x_test), len (y_test)))
```

# Fitting

```python
models = []
```

```python
def optimalKNN ():
    acc_manhat = []
    acc_euc = []
    neigh_man = 0
    neigh_euc = 0
    
    for i in range (1, 3):
        for j in range (1, 31): 
            model = KNeighborsClassifier (n_neighbors=j, p=i, metric='minkowski') 
            model.fit (x_train, y_train)
            y_pred = model.predict (x_test)
            if i == 1:
                acc_manhat.append (accuracy_score (y_test, y_pred))
            elif i == 2:
                acc_euc.append (accuracy_score (y_test, y_pred))
                
    visual_optimal (acc_manhat, acc_euc)
    
    print ("Accuracy (Manhattan): ", acc_manhat)
    print ("\nAccuracy (Euclidean): ", acc_euc)
    print ("\nMax Accuracy \nManhattan: {}\nEuclidean: {}".format (max (acc_manhat), max (acc_euc)))
    
    for i in range (len (acc_euc)):
        if (acc_euc [i] == 0.9824561403508771):
            print ('Euclidean Neighbor = {}'.format (i))
    
def visual_optimal (man, euc):
    plt.title ("Elbow point: Manhattan(Blue), Euclidean(Green)")
    plt.plot (man, 'b')
    plt.plot (euc, 'g')
    plt.show ()
```

```python
optimalKNN ()
```

```python
model_LR = LogisticRegression ()
model_DTC = DecisionTreeClassifier ()
model_RFC = RandomForestClassifier ()
model_KNN = KNeighborsClassifier (n_neighbors=10)
model_BNB = BernoulliNB ()
model_GNB = GaussianNB ()
model_MNB = MultinomialNB ()
model_SVC = SVC ()

models.extend ([model_LR, model_DTC, model_RFC, model_KNN, model_BNB, model_GNB, model_MNB, model_SVC])
print (models)
```

```python
trained_models = []

def train_models (lst):
    for model in lst:
        trained_models.append (model.fit (x_train, y_train))
        
    print (trained_models)

train_models (models)
```

# Accuracies

```python
def con_mat (lst):
    for model in lst:
        y_pred = model.predict (x_test)
        print (model)
        print (confusion_matrix (y_test, y_pred))
con_mat (trained_models)
```

```python
model_accuracies = []

def test_models (lst):
    for model in lst:
        y_pred = model.predict (x_test)
        model_accuracies.extend ([model, accuracy_score (y_test, y_pred), '|'])
        
test_models (trained_models)
print (model_accuracies)
```

```python
model_accuracies_only_int = [model_accuracies[i] for i in range (1, len (model_accuracies), 3)]
print (max (model_accuracies_only_int))
```

```python
for i in range (len (model_accuracies)):
    if model_accuracies [i] == 0.9766081871345029:
        print (model_accuracies [i-1])
```

```python
screen_size ['figure.figsize'] = 17, 7
plt.title ('Accuracies of models')
plt.xlabel ('Models')
plt.ylabel ('Accuracy')
plt.grid ()
plt.plot ([str (model_accuracies[i]) for i in range (0, len (model_accuracies), 3)], model_accuracies_only_int, 'g')
plt.savefig ('Model_Accuracy_Breast_Cancer.png')
plt.show ()
```

### Since Logistic Regression is giving us the highest accuracy (97.66081871345029 %) we will go with it for final model creation

```python
LR = LogisticRegression ()
LR.fit (x_train, y_train)
y_pred = LR.predict (x_test)
accuracy_score (y_test, y_pred)
```

```python
y_prediction = LR.predict (X)
accuracy_score (y, y_prediction)
y_prediction
```

## Our model accuracy is still ~ 95 % after predicting for whole dataset

### Thus we will be concluding Logistic Regression model as our final model
